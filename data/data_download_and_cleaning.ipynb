```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import random
from EA_funcs.utils import *
```

```python
# Time range to consider in the historical data
start = '2021-01-01'
end = '2024-04-05'

closing_prices = get_historical_data(tickers, start, end)

###### CETES ########
# Set the same number of closing prices (i.e. dates) for CETES as for other assets
num_days = len(closing_prices)
# Initialize with 1
values = [1]
for i in range(1, num_days):
    previous_value = values[i-1]
    new_value = previous_value * ((.1103/252) + 1)
    values.append(new_value)
    
# Create DataFrame for CETES
closing_prices_CETES = pd.DataFrame(values, columns=['CETES365'])
closing_prices_CETES.set_index(closing_prices.index, inplace=True)

## Combine closing price DataFrames
closing_prices = pd.concat([
    closing_prices,
    closing_prices_CETES
], axis=1)
```
